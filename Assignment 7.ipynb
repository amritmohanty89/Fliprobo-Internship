```python
# Preprocessing the data  
text = re.sub(r'\[[0-9]*\]',' ',paragraph)
text = re.sub(r'\s+',' ',text)

text = text.lower()

text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences
```

```python
sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sent_word
```

```python

for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i]
                    if word not in stopwords.words('english')]
```

```python
from sklearn.manifold import  TSNE
vocab = ['best','indian','e-commerce','company','founded','sellers','belong','new','incorporated','product','delivery','model','proven','great','significance','since','invested','infrastructure','electronics','website']
def tsne_plot(model):
    labels = []
    wordvecs = []

    for word in vocab:
        wordvecs.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=3, n_components=3, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = []
    y = []
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(15,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='left')

    plt.show()

tsne_plot(model)
```
